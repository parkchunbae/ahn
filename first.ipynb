## 유효컬럼
1. 성별(male,female)
2. 객실등급(1,2,3)
3. 출항지(  C, Q S)
4. 가족수(a+b)


```python
train.loc[train["Sex"] == "male", "Sex_encode"] = 0
train.loc[train["Sex"] == "female", "Sex_encode"] = 1

test.loc[test["Sex"] == "male", "Sex_encode"] = 0
test.loc[test["Sex"] == "female", "Sex_encode"] = 1

train.head()
```

```python
train.loc[train["Embarked"] == "C", "Embarked_encode"] = 0
train.loc[train["Embarked"] == "S", "Embarked_encode"] = 1
train.loc[train["Embarked"] == "Q", "Embarked_encode"] = 2

test.loc[test["Embarked"] == "C", "Embarked_encode"] = 0
test.loc[test["Embarked"] == "S", "Embarked_encode"] = 1
test.loc[test["Embarked"] == "Q", "Embarked_encode"] = 2
```

```python
feature_names = ['Pclass', 'Sex_encode', 'Embarked_encode']
label_name = "Survived"
```

```python
X_train = train[feature_names]
Y_train = train[label_name]


print(X_train.shape)

X_train.head()
Y_train.head()
```

```python
model = DecisionTreeClassifier(max_depth=5)
model
```

```python
model.fit(X_train, Y_train)
```
